El cálculo se basa en la suma de los tres elementos principales del costo de producción: Materia Prima Directa (MPD), Mano de Obra Directa (MOD) y Costos Indirectos de Fabricación (CIF).

\begin{equation*}
    \text{Costo de Producción} = \text{MPD} + \text{MOD} + \text{CIF}
\end{equation*}

- *Materia Prima Directa (MPD)*: Se calcula sumando el inventario inicial y las compras, y luego restando el inventario final. Los gastos de administración y comerciales no se incluyen en el cálculo del costo de producción.

\begin{equation*}
    \text{MPD} = \text{Inventario Inicial} + \text{Compras} - \text{Inventario Final}
\end{equation*}

- *Mano de Obra Directa (MOD)*: Se compone de los salarios netos y las cargas sociales.

\begin{equation*}
    \text{MOD} = \text{Salarios Netos} + \text{Cargas Sociales}
\end{equation*}

- *Costos Indirectos de Fabricación (CIF)*: Incluye todos los demás costos relacionados con la producción que no son MPD ni MOD.

\begin{equation*}
    \text{CIF} = \sum(\text{Costos Indirectos Individuales})
\end{equation*}

```python
def calcular_costo_de_produccion(inventario_inicial, compras, inventario_final, salarios_netos, cargas_sociales_mod, costos_indirectos_fabricacion):
    """
    Calcula el costo total de producción desglosado en sus tres elementos.
    
    Args:
        inventario_inicial (float): Valor del inventario inicial de materia prima.
        compras (float): Valor de las compras de materia prima del mes.
        inventario_final (float): Valor del inventario final de materia prima.
        salarios_netos (float): Costo de los salarios directos.
        cargas_sociales_mod (float): Costo de las cargas sociales directas.
        costos_indirectos_fabricacion (list): Lista de todos los costos indirectos.
    
    Returns:
        dict: Un diccionario con el costo de cada elemento y el costo total de producción.
    """
    # 1. Calcular Materia Prima Directa (MPD)
    mpd = inventario_inicial + compras - inventario_final
    
    # 2. Calcular Mano de Obra Directa (MOD)
    mod = salarios_netos + cargas_sociales_mod
    
    # 3. Calcular Costos Indirectos de Fabricación (CIF)
    cif = sum(costos_indirectos_fabricacion)
    
    # 4. Calcular el Costo Total de Producción
    costo_total = mpd + mod + cif
    
    return {
        "MPD": mpd,
        "MOD": mod,
        "CIF": cif,
        "Costo Total de Producción": costo_total
    }

# Datos del problema de Pérez SRL:
# Inventario de Acero (Materia Prima Directa)
inventario_inicial_acero = 72000
compras_mes_acero = 126000
inventario_final_acero = 108000

# Salarios y cargas sociales (Mano de Obra Directa)
salarios = 18000
cargas_sociales = 9000

# Lista de Costos Indirectos de Fabricación (CIF)
cif_lista = [
    900,   # Lubricantes
    1800,  # Depreciación y seguros
    3600,  # Combustibles
    5400,  # Sueldos y cargas sociales supervisores
    2700,  # Repuestos
    650,   # Seguros de producción
    3600   # Otros costos generales
]

# Gastos que NO son parte del costo de producción (para demostración)
gastos_operativos = {
    "Gastos comerciales": 3600,
    "Gastos de administracion": 2100
}

# Ejecutar la función con los datos de Pérez SRL
resultados = calcular_costo_de_produccion(
    inventario_inicial_acero,
    compras_mes_acero,
    inventario_final_acero,
    salarios,
    cargas_sociales,
    cif_lista
)

# Mostrar los resultados
print(f"Materia Prima Directa (MPD): ${resultados['MPD']:.2f}")
print(f"Mano de Obra Directa (MOD): ${resultados['MOD']:.2f}")
print(f"Costos Indirectos de Fabricación (CIF): ${resultados['CIF']:.2f}")
print("---")
print(f"Costo Total de Producción: ${resultados['Costo Total de Producción']:.2f}")
```

**Costo Total de Producción**

Finalmente, el costo total de producción es la suma de los tres componentes.

\begin{equation*}

\end{equation*}
