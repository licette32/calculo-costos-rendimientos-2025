### Funciones para los C√°lculos

Comenzaremos definiendo las funciones que se utilizar√°n en el cuaderno.

```python
# En esta celda definimos las funciones que usaremos en todo el notebook.

def calcular_cantidad_necesaria(unidades_a_producir, consumo_neto_por_unidad, porcentaje_merma):
    """
    Calcula la cantidad total de materia prima requerida, ajustando por la merma.
    """
    if porcentaje_merma >= 1:
        raise ValueError("El porcentaje de merma debe ser un valor decimal (ej. 0.20 para 20%).")
    
    material_neto = unidades_a_producir * consumo_neto_por_unidad
    cantidad_total = material_neto / (1 - porcentaje_merma)
    return cantidad_total

def calcular_costo_compra(cantidad_total_material, costo_por_unidad_material):
    """
    Calcula el costo total de la compra de materia prima.
    """
    return cantidad_total_material * costo_por_unidad_material
```

#### Ejercicio 1: Producto "ZZ"

Determinar la Cantidad necesaria de Materia Prima para una producci√≥n de 1.500 unidades del Producto ‚ÄúZZ‚Äù teniendo en cuenta:

- Consumo de Materia Prima por Unidad de Producto ‚ÄúZZ‚Äù................. 1‚ÅÑ2 litro
- Precio de la Materia Prima por litro................................. $200
- Merma de la Materia Prima ........................................... 20%

Para encontrar la cantidad total de material necesaria, primero se debe calcular la cantidad neta del material para la producci√≥n final. Luego, se ajusta esta cantidad para tener en cuenta el porcentaje de merma. 

La formula utilizada para este ajuste es: 

\begin{equation*}
Cantidad\ Total = \frac{Cantidad\ Neta}{1 - Porcentaje\ de\ Merma}
\end{equation*}

EL costo total es la cantidad de material necesaria multiplicada por su costo por unidad.

*Datos*:

- **Producci√≥n:** 1.500 unidades
- **Consumo:** 0,5 litro por unidad
- **Merma:** 20%
- **Precio:** $200 por litro

```python
cantidad_zz = calcular_cantidad_necesaria(1500, 0.5, 0.20)
costo_zz = calcular_costo_compra(cantidad_zz, 200)

print(f"Cantidad necesaria: {cantidad_zz:.2f} litros")
print(f"Costo total de compra: ${costo_zz:,.2f}")
```

\begin{equation*}
Cantidad\ Total\ (litros) = \frac{1500\ unidades \times 1,5\ \frac{litros}{unidad}}{1 - 0,20}
\end{equation*}

\begin{equation*}
Costo\ Total\ (\$) = 937,50\ litros \times \$200\ \frac{\$}{litro}
\end{equation*}

a) ¬øCu√°l es la cantidad inicial a incorporar de materia prima al proceso para la producci√≥n mensual?

b) Si el sector de compras informa que el costo por metro de materia prima es de $320. ¬øCu√°l es el importe en pesos a comprar para la producci√≥n del mes?

La f√≥rmula para calcular la cantidad total de materia prima necesaria, considerando la merma, es la siguiente:

\begin{equation*}
Cantidad\ Total = \frac{Cantidad\ Neta}{1 - Porcentaje\ de\ Merma}
\end{equation*}


Una vez que se tiene la cantidad total de material, el costo de compra se calcula multiplicando esta cantidad por el precio por unidad del material:

\begin{equation*}
Costo\ Total = Cantidad\ Total \times Costo\ por\ metro
\end{equation*}


*Datos*:

#### Ejercicio 4: Producto "A"

Determinar la Cantidad necesaria de Materia Prima para una producci√≥n de 2.000 unidades del Producto ‚ÄúA‚Äù teniendo en cuenta:

‚óã Consumo de Materia Prima por Unidad de Producto ‚ÄúA‚Äù..................500 grs.

‚óã Precio de la Materia Prima por kg...........................................................$10

‚óã Merma de la Materia Prima .....................................................................9%

Para calcular la cantidad necesaria y el costo de materia prima para una producci√≥n de 2000 unidades del Producto "A".

Se debe convertir el consumo de gramos a kilogramos antes de realizar los c√°lculos, ya que el precio se da por kg.

üßÆ *Datos*:

- **Producci√≥n**: 2.000 unidades 

- **Consumo neto por unidad**: 500 gramos (0.5 kg) 

- **Merma**: 9% (0.09) 

- **Precio por kg**: $10


\begin{equation*}
Cantidad\ Neta\ (kg) = 2000\ unidades \times 500\ \frac{grs}{unidad} \times \frac{1\ kg}{1000\ grs}
\end{equation*}

\begin{equation*}
Cantidad\ Total\ (kg) = \frac{Cantidad\ Neta\ (kg)}{1 - 0,09}
\end{equation*}

\begin{equation*}
Costo\ Total\ (\$) = 1098,90\ kg \times \$10\ \frac{\$}{kg}
\end{equation*}

```python
# Ejercicio 4: Producto "A"
unidades_a = 2000
consumo_neto_a_grs = 500
consumo_neto_a_kg = consumo_neto_a_grs / 1000  # Convertir gramos a kilogramos
merma_a = 0.09
precio_kg_a = 10

# Calcula la cantidad necesaria de material en kg.
cantidad_necesaria_a = calcular_cantidad_necesaria(unidades_a, consumo_neto_a_kg, merma_a)

# Calcula el costo total de la compra.
costo_total_a = calcular_costo_compra(cantidad_necesaria_a, precio_kg_a)

print(f"La cantidad necesaria de materia prima es: {cantidad_necesaria_a:.2f} kg")
print(f"El costo total de la compra es: ${costo_total_a:.2f}")
```
